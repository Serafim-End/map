```python
def apply_location(x):
    
    d = {}
    d['total_sum'] = x['total_cost'].sum()
    d['total_discount'] = x['total_discount'].sum()
    d['bill_count'] = x['set_id'].unique().size
    d['lng'] = x['lng'].values[0]
    d['lat'] = x['lat'].values[0]
    d['average_bill_price'] = float(d['total_sum']) / d['bill_count']
#     d['items'] = json.dumps([
#         (r.get('id'), ) for ind, r in x.iterrows()
#     ])
        
#     d['info'] = json.dumps({
#         r.get('set_id'): {
#             'total_cost': r.get('total_cost'),
#             'total_discount': r.get('total_discount'),
#             'item_name': r.get('item_name'),
#             'qnt': r.get('qnt'),
#             'price': r.get('price'),
#             'oper_discount': r.get('oper_discount'),
#             'id': r.get('id')
#         } for ind, r in x.iterrows()
#     })    
    return pd.DataFrame(data=d, index=x['loc_unique'])
```

```python
%%time
def apply_parallel(df_groupped, func):
    r = Parallel(n_jobs=multiprocessing.cpu_count())(delayed(func)(group) for name, group in df_groupped)
    return pd.concat(r)
df_points = apply_parallel(df_new.groupby(['loc_unique']), apply_location).drop_duplicates(subset=['lat', 'lng'])
```

```python
%%time

def func(d):
    
    a = []
    for ind, r in d.iterrows():
        lat = r.pop('lat')
        lng = r.pop('lng')
    
        a.append({
            'type': 'Feature',
            'geometry': {
                'type': 'Point',
                'coordinates': [lng, lat],
            },
            'properties': dict(zip(r.index, r.values))
        })
    return a


def parallelize_dataframe(df, func):
    
    core_a = []
    num_cores = multiprocessing.cpu_count() - 1
    num_partitions = num_cores
    df_split = np.array_split(df, num_partitions)
    pool = multiprocessing.Pool(num_cores)
    
    a = pool.map(func, df_split)
    core_a = sum(a, [])
    pool.close()
    pool.join()
    return core_a

core_features = parallelize_dataframe(df_points, func)
```
